```python

import os
import glob
import cv2
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import albumentations as A
import onnxruntime
```

```python
PATH = "C:/Users/moksh/segmentation/test_image/"

mask_files = glob.glob(PATH + '*_mask*')
image_files = [file.replace('_mask', '') for file in mask_files]

def diagnosis(mask_path):
    return 1 if np.max(cv2.imread(mask_path)) > 0 else 0

samples_df = pd.DataFrame({"image_path": image_files,
                  "mask_path": mask_files,
                  "diagnosis": [diagnosis(x) for x in mask_files]})

samples_df
```

```python
def load_sample(idx):
    original_image = cv2.imread(samples_df.iloc[idx, 0])
    original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    
    image = (np.array(original_image).astype(np.float32))/255.
    mask = cv2.imread(samples_df.iloc[idx, 1], 0)
    
    test_transform = A.Compose([
                    A.Resize(width=128, height=128, p=1.0)
                    ])
    
    aug = test_transform(image=image, mask=mask)
    image = aug['image']
    mask = aug['mask']
            
    image = image.transpose((2,0,1))
    
    #image normalize
    mean_vec = np.array([0.485, 0.456, 0.406])
    std_vec = np.array([0.229, 0.224, 0.225])
    for i in range(image.shape[0]):
        image[i, :, :] = (image[i, :, :] - mean_vec[i]) / (std_vec[i])
    
    mask = np.expand_dims(mask, axis=-1).transpose((2,0,1))
    return original_image, image, mask
```

```python
PATH = "C:/Users/moksh/segmentation/test_image/"
all_files = glob.glob(PATH + '*')
image_files = [file for file in all_files if '_mask' not in file]
mask_files = glob.glob(PATH + '*_mask*')

def diagnosis(mask_path):
    return 1 if os.path.exists(mask_path) and np.max(cv2.imread(mask_path)) > 0 else 0

samples_df = pd.DataFrame({
    "image_path": image_files,
    "mask_path": [file.replace('.jpg', '_mask.jpg') if file.replace('.jpg', '_mask.jpg') in mask_files else None for file in image_files],
    "diagnosis": [diagnosis(file.replace('.jpg', '_mask.jpg')) for file in image_files]
})

print(samples_df)
```

```python
def load_sample(idx):
    original_image = cv2.imread(samples_df.iloc[idx, 0])
    original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    
    image = (np.array(original_image).astype(np.float32))/255.
    
    test_transform = A.Compose([
                    A.Resize(width=128, height=128, p=1.0)
                    ])
    
    aug = test_transform(image=image)
    image = aug['image']
            
    image = image.transpose((2,0,1))
    
    #image normalize
    mean_vec = np.array([0.485, 0.456, 0.406])
    std_vec = np.array([0.229, 0.224, 0.225])
    for i in range(image.shape[0]):
        image[i, :, :] = (image[i, :, :] - mean_vec[i]) / (std_vec[i])
    
    mask_path = samples_df.iloc[idx, 1]
    if mask_path and os.path.exists(mask_path):
        mask = cv2.imread(mask_path, 0)
        mask = test_transform(image=mask)['image']
        mask = np.expand_dims(mask, axis=0)
    else:
        mask = None
    
    return original_image, image, mask
```

```python
pythonCopymodel_onnx = "C:/Users/moksh/segmentation/brain-mri-unet.onnx"
session = onnxruntime.InferenceSession(model_onnx, None)
input_name = session.get_inputs()[0].name
output_name = session.get_outputs()[0].name

def process_and_visualize(idx):
    original_img, test_img, test_mask = load_sample(idx)
    
    input_img = np.expand_dims(test_img, axis=0)
    result = session.run([output_name], {input_name: input_img})

    # Process the result
    arr = np.array(result).squeeze()
    arr = (arr > 0.5).astype(np.uint8)  # Apply threshold if needed

    # Visualize
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

    ax1.imshow(original_img)
    ax1.set_title('Original Image')
    ax1.axis('off')

    ax2.imshow(arr, cmap='gray')
    ax2.set_title('Segmentation Mask')
    ax2.axis('off')

    plt.tight_layout()
    plt.show()

    # Overlay
    resized_mask = cv2.resize(arr, (original_img.shape[1], original_img.shape[0]), interpolation=cv2.INTER_NEAREST)
    
    plt.figure(figsize=(15, 15))
    plt.imshow(original_img)
    plt.imshow(resized_mask, alpha=0.5, cmap='jet')
    plt.title('Original Image with Segmentation Overlay')
    plt.axis('off')
    plt.show()

    if test_mask is not None:
        print("Actual mask available")
        plt.figure(figsize=(10, 10))
        plt.imshow(test_mask.squeeze(), cmap='gray')
        plt.title('Actual Mask')
        plt.axis('off')
        plt.show()
    else:
        print("No actual mask available for this image")

# Process and visualize each image
for i in range(len(samples_df)):
    print(f"Processing image {i+1}/{len(samples_df)}")
    process_and_visualize(i)
```
